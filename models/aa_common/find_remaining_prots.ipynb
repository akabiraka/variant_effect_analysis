```python
import os
import sys
home_dir = "../../"
module_path = os.path.abspath(os.path.join(home_dir))
if module_path not in sys.path:
    sys.path.append(module_path)

import numpy as np
import pandas as pd
from models.aa_common.data_loader import get_protein_sequences
```

```python
popu_freq_protid_seq_dict = get_protein_sequences(home_dir, max_seq_len=1022, return_type="protid_seq_dict", data_type="popu_freq")
pathogenic_protid_seq_dict = get_protein_sequences(home_dir, max_seq_len=1022, return_type="protid_seq_dict", data_type="pathogenic")
likely_pathogenic_protid_seq_dict = get_protein_sequences(home_dir, max_seq_len=1022, return_type="protid_seq_dict", data_type="likely_pathogenic")
```

```python
popu_freq_protid_seq_set = set(popu_freq_protid_seq_dict.items())
pathogenic_protid_seq_set = set(pathogenic_protid_seq_dict.items())
likely_pathogenic_protid_seq_set = set(likely_pathogenic_protid_seq_dict.items())

remaining_from_pathogenic = pathogenic_protid_seq_set - popu_freq_protid_seq_set
print(len(remaining_from_pathogenic))

remaining_from_likely_pathogenic = likely_pathogenic_protid_seq_set - popu_freq_protid_seq_set
print(len(remaining_from_likely_pathogenic))

total_remaining = remaining_from_pathogenic.union(remaining_from_likely_pathogenic)
print(len(total_remaining))
```

```python
out_fasta_filepath = home_dir+"models/aa_common/merged_np_nm_seq/remaining_sequences.fasta"
if os.path.exists(out_fasta_filepath): os.remove(out_fasta_filepath)
out_fasta_file_handle = open(out_fasta_filepath, "a")

for (protid, seq) in total_remaining:
    print(protid, seq)
    out_fasta_file_handle.write(f">{protid}\n")
    out_fasta_file_handle.write(f"{seq}\n")
    # break
out_fasta_file_handle.close() 
```
