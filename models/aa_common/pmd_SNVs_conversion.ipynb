```python
import os
import sys
home_dir = "../../"
module_path = os.path.abspath(os.path.join(home_dir))
if module_path not in sys.path:
    sys.path.append(module_path)


import pandas as pd
from utils.pandas_extented_filters import filter_remove_null_nan_empty_entries
```

```python
pmd_df = pd.read_csv(home_dir+"data/pmd/dfSNP_PMD.csv", sep=",") # PMD: protein mutation dataset
print(pmd_df.shape) # (55465, 54)
print(pmd_df.columns)
```

```python
columns = ['mut_id', 'pmd_id', 'protein', 'mut_PMD', 'mut_real', 'function', 'taxid', 'function_e', 'function_e2', 'functional_effect', 'seq']
# function_e - ranks the effect from 0 to 5 (no effect to knock-out). 
#   0: no-effect, 1-4: effect, 5: knock-out (means no function at all)
# functional_effect (categorical: no effect, effect, knock-out)
pmd_selected_cols_df = pmd_df[columns]

#  mut_id, pmd_id are very clean column, no nan, null, or empty string found
a_col_name = "mut_PMD" # mut_id, pmd_id,
print(pmd_selected_cols_df[pd.isna(pmd_df[a_col_name])].shape)
print(pmd_selected_cols_df[pd.isnull(pmd_df[a_col_name])].shape)
print(pmd_selected_cols_df[pmd_selected_cols_df[a_col_name] == ""])


pmd_removed_NNE_df = filter_remove_null_nan_empty_entries(pmd_selected_cols_df, a_col_name="mut_PMD") # NNE: null, nan, empty
print(f"After removing NNE (null, nan and empty) values: {pmd_removed_NNE_df.shape}") # (55465, 11)
# pmd_removed_NNE_and_knockout_df = pmd_removed_NNE_df[pmd_removed_NNE_df["functional_effect"] != 'Knock-out']
# print(f"After removing 'Knock-out' mutations: {pmd_removed_NNE_and_knockout_df.shape}") # (16244, 11)

print(pmd_removed_NNE_df["functional_effect"].value_counts()) # No effect: 8680, Effect: 7564
pmd_removed_NNE_df
```

### Generating corresponding fasta file

```python
# unique_pmdid_seq_df = pmd_removed_NNE_and_knockout_df[["pmd_id", "protein", "seq"]].drop_duplicates()
# print(unique_pmdid_seq_df["pmd_id"].unique().shape)
# print(unique_pmdid_seq_df["pmd_id"].value_counts())
# print(unique_pmdid_seq_df[unique_pmdid_seq_df["pmd_id"]=="A911139"]) # same pmd_id has multiple proteins
import uuid

unique_protein_seq_df = pmd_removed_NNE_df[["protein", "seq"]].drop_duplicates()
print(unique_protein_seq_df.shape)
for tuple in unique_protein_seq_df.itertuples():
    unique_protein_seq_df.loc[tuple.Index, "protein_id"] = "protid_"+str(uuid.uuid1().fields[0])
print(unique_protein_seq_df.shape)
print(unique_protein_seq_df["protein_id"].unique().shape)


pmd_withgenerated_protid_df = pmd_removed_NNE_df.merge(unique_protein_seq_df, how="left", on=["protein", "seq"])
print(pmd_withgenerated_protid_df["protein_id"].unique().shape)


out_fasta_filepath = home_dir+"models/aa_common/datasets_pmd_analysis/pmd_sequences.fasta"
if os.path.exists(out_fasta_filepath): os.remove(out_fasta_filepath)
out_fasta_file_handle = open(out_fasta_filepath, "a")

for i, tuple in enumerate(unique_protein_seq_df.itertuples()):
    # print(tuple.pmd_id, tuple.seq)
    out_fasta_file_handle.write(f">{tuple.protein_id} {tuple.protein}\n")
    out_fasta_file_handle.write(f"{tuple.seq}\n")
    # break
out_fasta_file_handle.close()    

pmd_withgenerated_protid_df
```

### Generating variants file

```python
pmd_withgenerated_protid_df["wt"] = pmd_withgenerated_protid_df["mut_real"].apply(lambda x: x[0]) # 1-letter amino acid
pmd_withgenerated_protid_df["mut"] = pmd_withgenerated_protid_df["mut_real"].apply(lambda x: x[-1])
pmd_withgenerated_protid_df["prot_pos"] = pmd_withgenerated_protid_df["mut_real"].apply(lambda x: int(x[1:-1])) # mutation position is 1-indexed
pmd_withgenerated_protid_df.to_csv(home_dir+"models/aa_common/datasets_pmd_analysis/pmd_data.csv", sep=",", index=False, header=True)
pmd_withgenerated_protid_df
```
